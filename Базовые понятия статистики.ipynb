Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
df=pd.read_csv('/content/horse_data.csv', na_values='?')
df
```

```python
df = df[['1', '2', '4','5','6','7', '11', '23']]
df
```

```python
df.columns = ['surgery', 'age', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'pain', 'outcome']
```

```python
df
```

```python
df.info()
```

```python
#количество пропусков (больше всего в rectal_temperature , respiratory_rate, temperature_of_extremities, pain )
df.isna().mean()
```

```python
#базовые статистики
df.describe()
```

```python
#выбросы для колонки age. 
```

```python
q1=df.age.quantile(0.25)
q3=df.age.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
df[~df['age'].between(lower,upper, inclusive=True)]['age'].value_counts()
```

```python
#возраст - либо 1, либо 9. Согласно описания возраст может быть либо 1 - взрослая лошадь, либо 2 - молодая. 
# 9 - это какие-то ошибочные записи - выбросы
df.age.value_counts()
```

```python
#выбросы для колонки rectal_temperature, нормальная температура - 37.8 (границы выбросов - 36.6-39.6)
q1=df.rectal_temperature.quantile(0.25)
q3=df.rectal_temperature.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
df[~df['rectal_temperature'].between(lower,upper, inclusive=True)]['rectal_temperature'].value_counts()
```

```python
#выбросы для колонки pulse. Норма - это 30-40. Все, что выше 150 - относим к выбросам
q1=df.pulse.quantile(0.25)
q3=df.pulse.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
df[~df['pulse'].between(lower,upper, inclusive=True)]['pulse'].value_counts()
```

```python
#выбросы для колонки respiratory_rate
#полезность сомнительна из-за больших колебаний
q1=df.respiratory_rate.quantile(0.25)
q3=df.respiratory_rate.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
df[~df['respiratory_rate'].between(lower,upper, inclusive=True)]['respiratory_rate'].value_counts()
```

```python
# в поле surgery пропущено 1 значение
df[df.surgery.isna()]
```

```python
df[df.outcome.isna()]
```

```python
#заполним пропущеные значения медианным значением
```

```python
df['surgery']=df['surgery'].fillna(df['surgery'].median())
```

```python
df['rectal_temperature']=df['rectal_temperature'].fillna(df['rectal_temperature'].median())
```

```python
df['pulse']=df['pulse'].fillna(df['pulse'].median())
```

```python
df['respiratory_rate']=df['respiratory_rate'].fillna(df['respiratory_rate'].median())
```

```python
df['temperature_of_extremities']=df['temperature_of_extremities'].fillna(df['temperature_of_extremities'].median())
```

```python
df['pain']=df['pain'].fillna(df['pain'].median())
```

```python
df['outcome']=df['outcome'].fillna(df['outcome'].median())
```
